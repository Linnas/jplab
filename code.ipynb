```python
from scipy.special import xlogy
```

```python
from scipy.stats import entropy
```

```python
from scipy.spatial.distance import jensenshannon
```

```python
def diversity(arr):
    return entropy(list(Counter(arr).values()))
```

```python
def divergence(arr1,arr2):
    counter1={}
    counter2={}
    for i in set(arr1)|set(arr2):
        counter1[i]=0
        counter2[i]=0
    print(counter1)
    Counter1=dict(Counter(arr1))
    print(Counter1)
    for i in set(arr1):
        counter1[i]=Counter1[i]
    print(counter1)
    Counter2=dict(Counter(arr2))
    for i in set(arr2):
        counter2[i]=Counter2[i]
    return jensenshannon(list(counter1.values()),list(counter2.values()))
```

下面是一个例子，假设随机把这些细胞分成4类
即前端需传回三个参数：cluster、chain以及col
这里cluster取随机数0-3，chain取TRA，col取v_gene
返回cluster名字（对应前端plotly括号外的值），diversity结果（对应前端plotly括号内的值）和divergence(对应前端plotly数据中的z）

```python
clusterRandom=np.random.randint(0,4,size=len(cells))
chain='TRA'
col='v_gene'
```

```python
def calculation(dataframe,cluster,chain,col):
    cells=dataframe.index.unique()
    cell2cluster=pd.DataFrame(cluster,index=cells,columns=['cluster'])
    df2['cluster']=cell2cluster.loc[df2.index]
    df2sub=df2.loc[df2['chain']==chain]
    data=[]
    clusterName=[]
    for i in df2sub.groupby(by='cluster')[col]:
        clusterName.append(i[0])
        data.append(i[1].values)

    numClusters=len(np.unique(cluster))

    diversities=np.zeros(numClusters)
    divergences=np.zeros([numClusters,numClusters])

    for i in range(0,numClusters):
        diversities[i]=diversity(data[i])
        for j in range(0,numClusters):
            divergences[i,j]=divergence(data[i],data[j]) if i!=j else 0
    
    return clusterName,diversities,divergences
```
